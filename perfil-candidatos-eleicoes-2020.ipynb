# Perfil dos Candidatos às Eleições 2020

### Importação das bibliotecas necessárias

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

### Importação dos dados e seleção das colunas

```python
data = pd.read_csv('/home/evandro/projetos/perfil-candidatos-eleicoes-2020/consulta_cand_2020_SP.csv', sep=';', encoding='ISO-8859-1')
```

```python
required_columns = ['DS_CARGO', 'CD_GENERO', 'DS_GENERO', 'CD_GRAU_INSTRUCAO', 'DS_GRAU_INSTRUCAO']
```

```python
female_candidates = []
male_candidates = []
educational_level_read = []
educational_level_read_write = []
educational_level_elementary_inc = []
educational_level_elementary_comp = []
educational_level_high_school_inc = []
educational_level_high_school_comp = []
educational_level_university_inc = []
educational_level_university_comp = []
```

```python
data.head()
```

```python
required_data = data.filter(items=required_columns)
```

## Total de Candidatos por Gênero

```python
female_candidates.append(data.query('(CD_GENERO == 4)')['CD_GENERO'].count())
male_candidates.append(data.query('(CD_GENERO == 2)')['CD_GENERO'].count())
```

```python
total_candidates = np.array(male_candidates[0]) + np.array(female_candidates[0])

x1 = [str('Feminino')]
x2 = [str('Masculino')]
y1 = [(int(female_candidates[0]))]
y2 = [(int(male_candidates[0]))]

plt.title('Candidatos por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Candidatos')

plt.bar(x1, y1, label = 'Mulheres', color='#5DB7A1')
plt.bar(x2, y2, label = 'Homens', color='#00F3B9')
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

## Total de candidatos por Grau de Instrução

```python
educational_level_read.append(data.query('(CD_GRAU_INSTRUCAO == 1)')['CD_GRAU_INSTRUCAO'].count())
educational_level_read_write.append(data.query('(CD_GRAU_INSTRUCAO == 2)')['CD_GRAU_INSTRUCAO'].count())
educational_level_elementary_inc.append(data.query('(CD_GRAU_INSTRUCAO == 3)')['CD_GRAU_INSTRUCAO'].count())
educational_level_elementary_comp.append(data.query('(CD_GRAU_INSTRUCAO == 4)')['CD_GRAU_INSTRUCAO'].count())
educational_level_high_school_inc.append(data.query('(CD_GRAU_INSTRUCAO == 5)')['CD_GRAU_INSTRUCAO'].count())
educational_level_high_school_comp.append(data.query('(CD_GRAU_INSTRUCAO == 6)')['CD_GRAU_INSTRUCAO'].count())
educational_level_university_inc.append(data.query('(CD_GRAU_INSTRUCAO == 7)')['CD_GRAU_INSTRUCAO'].count())
educational_level_university_comp.append(data.query('(CD_GRAU_INSTRUCAO == 8)')['CD_GRAU_INSTRUCAO'].count())
```

```python
x1 = [str('1')]
x2 = [str('2')]
x3 = [str('3')]
x4 = [str('4')]
x5 = [str('5')]
x6 = [str('6')]
x7 = [str('7')]
x8 = [str('8')]

y1 = [(int(educational_level_read[0]))]
y2 = [(int(educational_level_read_write[0]))]
y3 = [(int(educational_level_elementary_inc[0]))]
y4 = [(int(educational_level_elementary_comp[0]))]
y5 = [(int(educational_level_high_school_inc[0]))]
y6 = [(int(educational_level_high_school_comp[0]))]
y7 = [(int(educational_level_university_inc[0]))]
y8 = [(int(educational_level_university_comp[0]))]

plt.title('Candidatos por Grau de Instrução')
plt.xlabel('Grau de Instrução')
plt.ylabel('Número de Candidatos')

plt.bar(x1, y1, label = '1 - Lê', color='#75B0B4')
plt.bar(x2, y2, label = '2 - Lê e Escreve', color='#7595B4')
plt.bar(x3, y3, label = '3 - Ensino Fundamental Incompleto', color='#8975B4')
plt.bar(x4, y4, label = '4 - Ensino Fundamental Completo', color='#C2727C')
plt.bar(x5, y5, label = '5 - Ensino Médio Incompleto', color='#E14F45')
plt.bar(x6, y6, label = '6 - Ensino Médio Completo', color='#EF8C54')
plt.bar(x7, y7, label = '7 - Superior Incompleto', color='#EFC454')
plt.bar(x8, y8, label = '8 - Superior Completo', color='#D9EF54')
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
labels = ['Lê e Escreve',
          'Ensino Fundamental Incompleto',
          'Ensino Fundamental Completo',
          'Ensino Médio Incompleto',
          'Ensino Médio Completo',
          'Superior Incompleto',
          'Superior Completo']

sizes = [ 
    int(educational_level_read_write[0]),
    int(educational_level_elementary_inc[0]),
    int(educational_level_elementary_comp[0]),
    int(educational_level_high_school_inc[0]),
    int(educational_level_high_school_comp[0]),
    int(educational_level_university_inc[0]),
    int(educational_level_university_comp[0])]

explode = (0, 0, 0, 0, 0.1, 0, 0)

fg1, ax1 = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    
    return '{:.1f}%'.format(pct, absolute)

colors = ['#7595B4','#8975B4','#C2727C','#E14F45','#EF8C54','#EFC454','#D9EF54']

wedges, texts, autotexts = ax1.pie(sizes, explode=explode, labels=labels, shadow=True, startangle=90, colors=colors, autopct=lambda pct: func(pct, sizes), textprops=dict(color='w'))


ax1.legend(wedges, labels, 
          title='Grau de Instrução',
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=12, weight='bold')

ax1.axis('equal')

ax1.set_title('Grau de Instrução dos Candidatos às eleições 2020')

plt.show()
```
